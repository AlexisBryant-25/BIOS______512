## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

A patch represents the difference between 2 different versions of a file. So, the patch is the change itself and when you commit you are recording the change in history.

## Question 3
#### What is the difference between a branch and a fork?

A branch is a pointer or internal divergence to certain commits within the same repository while a fork is an external copy of a repository under a different user.

## Question 4
#### How are diff and patch related? What’s the difference?

Diff is a git command that displays the differences between 2 versions of a file or commit. A patch applies the differences to transform the first version into the second version of the file. So diff and patch work together in the way that diff describes the changes while patch implements the changes.
