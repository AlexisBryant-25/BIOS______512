## Question 1
This question will walk you through creating your own `kmeans` function.

#### a) What are the steps of `kmeans`?
**Hint**: There are 4 steps/builder functions that you'll need.

The four steps of kmeans are: 
1. Randomly assign each observation to one of the clusters
2. Compute cluster means (center points)
3. Reassign the points to the nearest mean
4. Recompute the means until convergence (cluster assignments stop changing)


#### b) Create the builder function for step 1.

```r
library(tidyverse)

initialize_clusters <- function(data, k) {
    labels <- sample(1:k, nrow(data), replace = TRUE)
    return(labels)
}
    
```

#### c) Create the builder function for step 2.

```r
compute_means <- function(data, labels, k) {
    means <- matrix(NA, nrow = k, ncol = ncol(data))
    for(i in 1:k) {
        means[i, ] <- colMeans(data[labels == i, , drop = FALSE])
    }
    return(means)
}
```

#### d) Create the builder function for step 3.
*Hint*: There are two ways to do this part - one is significantly more efficient than the other. You can do either.  

```r
assign_cluster_vec <- function(data, means) {
    distances <- as.matrix(dist(rbind(means, data)))[1:nrow(means), (nrow(means) + 1):(nrow(means) + nrow(data))]
    labels <- apply(distances, 2, which.min)
    return(labels)
    }
```

#### e) Create the builder function for step 4.

```r
kmeans_done <- function(old_labels, new_labels) {
    all(old_labels == new_labels)
    }
```

#### f) Combine them all into your own `kmeans` function.

```r
my_kmeans <- function(data, k, max_iter = 100) {
    labels <- initialize_clusters(data, k)

    for(iter in 1:max_iter) {
        means <- compute_means(data, labels, k)
        new_labels <- assign_cluster_vec(data, means)

        if(kmeans_done(labels, new_labels)) {
            cat("Converged after", iter, "iterations\n")
            break
        }
        return(list(labels = labels, means = means))
    }
    }
```

## Question 3
#### a) Explain the process of using a for loop to assign clusters for kmeans.

In using a loop to assign clusters you iterate over each observation i.
First, you compute the euclidean distance to each cluster center. Then, find the minimum distance among all the clusters and assign the datapoint to the cluster with the smallest distance and repeat until all the points are assigned a cluster. Using the loop can be very slow because it performs the distance calculations one at a time.

#### b) Explain the process of vectorizing the code to assign clusters for kmeans.

Vectorizing the code helps to be able to do the same operation on all the data points at the same time by using matrix calculations so you don't have to loop through each point. To do this you create a function that combines the means and data to compute all the distances at once. In the distance calculation you use "as.matrix" at the beginning so it returns a distance matrix between all points and centroids.

#### c) State which (for loops or vectorizing) is more efficient and why.

Vectorizing is more efficient because it is faster than iterating over every point, requires less lines of code, and when working with larger datasets it is more scalable and practical.

## Question 4
#### When does `kmeans` fail? What assumption does `kmeans` use that causes it to fail in this situation?

Kmeans fails when the clusters in the data are not equally sized and spherical. Kmeans assumes all clusters have similar variance and distance and are distributed in euclidean space. Using this assumption if the clusters are of variying sizes kmeans will misclassify many of the points an lead to incorrct results due to varying mean distances.

## Question 5
#### What assumption do Guassian mixture models make?

Gaussian mixture models assume that data is drawn from several distributions with a mean and covariance matrix. So clusters can have different shapes and sizes, which means that points can be in multiple clusters adding flexibility in modeling more complex datasets.

## Question 6
#### What assumption does spectral clustering make? Why does this help us?

Spectral clustering assumes that close points belong in the same cluster. This is not a strong assumption because it relies only on the graph of pairwise similarities but it is useful in handling clustering with clusters that have unsusual shapes and nonlinear data.

#### Define the gap statistic method. What do we use it for?

The gap statistic method compares the total within lcuster variation for different values of k to a random baseline. Its useful in identifying the point where adding more clusters would not be beneficial. We use it to objectively find k instead of trial and error.
